```python
import pandas as pd
import matplotlib
from matplotlib.dates import DateFormatter
import datetime as DT
%matplotlib widget

df = pd.read_csv("orbit_early_harvest_orbits.txt",header=None,delimiter=" ", names=["Date","SatLat","SatLon","SatLT","LPLat","LPLon","LPLT","SZA","SSA"])
df['Date']= pd.to_datetime(df['Date'])
df['LPLT']= pd.to_datetime(df['LPLT'],format='%H:%M:%S')
df['SatLT']= pd.to_datetime(df['SatLT'],format='%H:%M:%S')
startdate = DT.datetime(2022,11,9,12,48,0)
enddate = DT.datetime(2022,11,9,14,10,0)

df = df[(df.Date < enddate)]
df = df[(startdate < df.Date)]
```

```python
ax = df.plot.scatter("Date","LPLat")
date_form = DateFormatter("%H:%M")
ax.xaxis.set_major_formatter(date_form)
```

```python
ax = df.plot.scatter("LPLT","LPLat")
date_form = DateFormatter("%H:%M")
ax.xaxis.set_major_formatter(date_form)
```
